```python
from ultralytics import YOLO\nmodel = YOLO('yolo11n.pt')\nresults = model.train(data='/content/road_insight/ml/dataset/pothole.yaml', epochs=100, imgsz=640, batch=16, patience=20, project='road_insight', name='yolo11n_pothole_v1', close_mosaic=10)\n
```

```python
metrics = model.val(data='/content/road_insight/ml/dataset/pothole.yaml', split='test', conf=0.6, iou=0.5)\nprint(metrics)\n
```
